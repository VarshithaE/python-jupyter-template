```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = x**2

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$y = x^2$', color='b', linewidth=2)

# Customization
plt.title('Square Function Plot', fontsize=14)
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values (Square of X)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']

# Plot
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color=colors)

# Customization
plt.title('Category Value Distribution', fontsize=14)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show plot
plt.show()
```

```python

import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='blue', marker='o')

# Annotations
for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=10, ha='right', va='bottom')

# Customization
plt.title('Scatter Plot with Annotations', fontsize=14)
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)
y = x**2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot
axs[0, 0].plot(x, y, color='blue', linewidth=2)
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[0, 0].grid(True)

# Bar Plot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.6)

# Scatter Plot
axs[1, 0].scatter(x_scatter, y_scatter, color='blue', marker='o')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')
axs[1, 0].grid(True, linestyle='--', alpha=0.6)

# Pie Chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'purple'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Plot
plt.figure(figsize=(6, 4))
plt.hist(data, bins=5, color='blue', edgecolor='black', alpha=0.7)

# Customization
plt.title('Histogram of Data', fontsize=14)
plt.xlabel('Values', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show plot
plt.show()
```
